# Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

The Probability Mass Function (PMF) and Probability Density Function (PDF) are both ways to describe the probability distribution of a random variable.

The PMF is used for discrete random variables, where the possible outcomes are countable, and it gives the probability that the random variable takes on a specific value. The PMF is defined as:

P(X = x) = probability that the random variable X takes on the value x

For example, suppose you have a fair six-sided die. The PMF for this die would be:

P(X = 1) = 1/6
P(X = 2) = 1/6
P(X = 3) = 1/6
P(X = 4) = 1/6
P(X = 5) = 1/6
P(X = 6) = 1/6

This means that the probability of rolling a 1 on the die is 1/6, the probability of rolling a 2 is 1/6, and so on.

On the other hand, the PDF is used for continuous random variables, where the possible outcomes are uncountable, and it gives the probability density at a specific point in the distribution. The PDF is defined as:

f(x) = probability density at x

The probability of the random variable taking on a specific value is then the area under the PDF curve at that point.

For example, suppose you have a normal distribution with mean 0 and standard deviation 1. The PDF for this distribution is:

f(x) = (1/sqrt(2*pi)) * e^(-x^2/2)

This means that the probability density at x=0 is 0.3989, but the probability of the random variable taking on the value x=0 is actually 0, since the area under the curve at that point is infinitesimally small. Instead, the probability of the random variable being within some range of values is found by integrating the PDF over that range. For example, the probability of the random variable being between -1 and 1 is approximately 0.6827, since the area under the curve between those points is 68.27% of the total area under the curve.




For the PMF of a fair six-sided die:

```python
import numpy as np

# Define the PMF
pmf = np.ones(6) / 6

# Print the PMF
print("PMF:", pmf)
```

For the PDF of a normal distribution with mean 0 and standard deviation 1:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the PDF
def pdf(x):
    return 1 / np.sqrt(2 * np.pi) * np.exp(-x**2 / 2)

# Define the range of x values
x = np.linspace(-4, 4, 1000)

# Compute the PDF for each x value
y = pdf(x)

# Plot the PDF
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("PDF")
plt.show()
```

# Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

The Cumulative Density Function (CDF) is a function used in probability theory to describe the probability distribution of a random variable. The CDF of a random variable X is defined as:

F(x) = P(X ≤ x)

That is, it gives the probability that the random variable X takes on a value less than or equal to x.

The CDF is useful because it allows us to compute the probability of a random variable being within a certain range of values. Specifically, the probability that the random variable X is between a and b is given by:

P(a ≤ X ≤ b) = F(b) - F(a)

Here's an example to illustrate how the CDF works:

Suppose we have a random variable X that follows a standard normal distribution, with mean 0 and standard deviation 1. The CDF of X is given by:

F(x) = P(X ≤ x) = ∫_-∞^x (1/√(2π)) * e^(-t^2/2) dt

To calculate the probability that X is between -1 and 1, we can use the CDF:

P(-1 ≤ X ≤ 1) = F(1) - F(-1) = P(X ≤ 1) - P(X ≤ -1)

Using a standard normal distribution table or a calculator, we can find that P(X ≤ 1) ≈ 0.8413 and P(X ≤ -1) ≈ 0.1587. Therefore:

P(-1 ≤ X ≤ 1) ≈ 0.8413 - 0.1587 = 0.6826

This means that there is a 68.26% chance that X falls within the range of -1 to 1.

The CDF is used in many areas of statistics and probability, such as hypothesis testing, confidence intervals, and simulation studies. It provides a convenient way to compute probabilities for a wide variety of probability distributions, including those that don't have a simple closed-form expression for their probability density or mass functions.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the parameters of the normal distribution
mu = 0
sigma = 1

# Define the range of x values
x = np.linspace(-4, 4, 1000)

# Compute the CDF for each x value
y = norm.cdf(x, mu, sigma)

# Plot the CDF
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("CDF")
plt.show()
```

# Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?

The Bernoulli distribution is a discrete probability distribution that models a single trial of a binary experiment. In other words, it models the probability of success or failure for a single event. The Bernoulli distribution is named after Swiss mathematician Jakob Bernoulli.

The probability mass function (PMF) of the Bernoulli distribution is given by:

P(X=1) = p

P(X=0) = 1-p

where X is the random variable, p is the probability of success, and (1-p) is the probability of failure.

An example of a Bernoulli distribution is flipping a coin, where the probability of getting heads is p and the probability of getting tails is (1-p). Another example is rolling a dice and getting a specific number.

The Bernoulli distribution is different from the binomial distribution in that the Bernoulli distribution models a single trial of a binary experiment, while the binomial distribution models multiple trials of a binary experiment.

The binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent trials. Each trial has only two possible outcomes, success or failure. The probability of success (p) is the same for each trial. The number of trials is denoted by n, and the number of successes is denoted by x.

The probability mass function (PMF) of the binomial distribution is given by:

P(X=x) = (n choose x) * p^x * (1-p)^(n-x)

where (n choose x) is the binomial coefficient, which represents the number of ways to choose x successes from n trials.

An example of a binomial distribution is flipping a coin multiple times and counting the number of heads. Another example is conducting a survey and counting the number of people who answer "yes" to a specific question.

In summary, the Bernoulli distribution models a single trial of a binary experiment, while the binomial distribution models multiple trials of a binary experiment. The Bernoulli distribution has only one parameter (p), while the binomial distribution has two parameters (p and n).

# Q7. Explain uniform Distribution with an example.

Uniform distribution is a probability distribution where all possible outcomes are equally likely. In other words, it is a distribution where the probability of each outcome within a certain range is constant. A uniform distribution can be discrete or continuous.

A discrete uniform distribution assigns equal probability to a finite number of outcomes. For example, rolling a fair six-sided dice has a discrete uniform distribution since each of the six possible outcomes (1, 2, 3, 4, 5, or 6) has an equal chance of occurring.

A continuous uniform distribution assigns equal probability to any interval of the same length within a certain range. For example, the height of people between 5 feet and 6 feet has a continuous uniform distribution since any height within this range is equally likely.

In general, the probability density function of a continuous uniform distribution can be written as:

f(x) = 1 / (b - a)

where a is the minimum possible value and b is the maximum possible value.

An example of a continuous uniform distribution is the distribution of the arrival time of a bus, assuming that the bus arrives between 8:00 am and 9:00 am with equal probability at any time within this range.





```python
import numpy as np

a = 5  # minimum value
b = 10  # maximum value

random_numbers = np.random.uniform(a, b, size=10)
print(random_numbers)
```

# Q8. What is the z score? State the importance of the z score.

The z-score, also known as the standard score, is a measure of how many standard deviations an observation or data point is from the mean of its distribution. It is calculated by subtracting the mean from the observation and dividing the result by the standard deviation:

z = (x - μ) / σ

where z is the z-score, x is the observation, μ is the mean, and σ is the standard deviation.

The z-score is important because it allows us to compare observations from different normal distributions on the same scale. By converting observations to their corresponding z-scores, we can determine how unusual or typical a particular observation is compared to other observations in its distribution.

The z-score is also useful for hypothesis testing, as it allows us to calculate the probability of observing a particular value or a more extreme value, assuming that the data is normally distributed.

Furthermore, the z-score is used in the construction of confidence intervals, where we want to estimate the range of values that a population parameter is likely to fall within, given a sample from the population.

In summary, the z-score is an important tool in statistics that allows us to standardize observations and compare them across different distributions, as well as to conduct hypothesis testing and construct confidence intervals.





# Q9. What is Central Limit Theorem? State the significance of the Central Limit Theorem.

The Central Limit Theorem (CLT) is a fundamental concept in statistics that states that the sampling distribution of the mean of any independent, identically distributed random variables will tend to follow a normal distribution, as the sample size increases, regardless of the shape of the original distribution.

In other words, if we take multiple samples from a population and calculate the mean of each sample, the distribution of those sample means will be approximately normal, even if the original population distribution is not normal.

The significance of the Central Limit Theorem lies in its usefulness in inferential statistics. It allows us to make statistical inferences about a population parameter, such as the population mean or proportion, based on a sample of observations, even when the population distribution is unknown or not normally distributed.

For example, we can use the Central Limit Theorem to construct confidence intervals for a population mean or to perform hypothesis testing on population parameters, even if the underlying data is not normally distributed. This is because the normal distribution is well-understood and has many useful properties that can be used to make statistical inferences.

Overall, the Central Limit Theorem is an important concept in statistics that allows us to use the normal distribution to make statistical inferences about population parameters, even when the population distribution is unknown or not normally distributed.

> The Central Limit Theorem (CLT) relies on a few key assumptions in order to hold:

- 1.Random sampling: The sample must be drawn randomly from the population of interest. This means that each observation in the population has an equal chance of being selected for the sample.

- 2.Independent observations: The observations in the sample must be independent of each other. This means that the value of one observation should not depend on the value of any other observation in the sample.

- 3.Sample size: The sample size should be large enough. While there is no fixed threshold for what constitutes a "large enough" sample size, a commonly cited rule of thumb is that the sample size should be at least 30. However, in practice, the required sample size can depend on the shape of the population distribution.

- 4.Finite variance: The population distribution from which the sample is drawn must have a finite variance. This means that the variance of the population must not be infinite.

If these assumptions are met, the Central Limit Theorem holds, and we can make useful inferences about the population distribution based on the sample mean. If these assumptions are not met, the Central Limit Theorem may not hold, and alternative methods may be required to make inferences about the population distribution.
